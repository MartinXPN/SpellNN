```python
from collections import Counter
import operator

from tqdm import tqdm_notebook as tqdm
import matplotlib.pyplot as plt
%matplotlib inline

import sys
sys.path.append('..')
```

```python
def show_dataset_statistics(path, max_lines=10):
    
    paragraph_lengths = []
    paragraph_words = []
    word_lengths = []
    
    with open(path, 'r') as f:
        for paragraph in tqdm(f):
            paragraph = paragraph.strip()
            if paragraph == '':
                continue
            
            paragraph_lengths.append(len(paragraph))
            words = paragraph.split()
            paragraph_words.append(len(words))
            word_lengths += [len(w) for w in words]
            
            max_lines -= 1
            if max_lines == 0:
                break
    
    f, ax = plt.subplots(ncols=3)
    f.set_figheight(3)
    f.set_figwidth(20)
    f.dpi = 150
    
    # Word lengths
    ax[0].hist(word_lengths,      facecolor='b', bins=50); ax[0].set_xlabel('Word lengths');     ax[0].set_ylabel('Counts')
    ax[1].hist(paragraph_lengths, facecolor='r', bins=50); ax[1].set_xlabel('Paragraph length'); ax[1].set_ylabel('Counts')
    ax[2].hist(paragraph_words,   facecolor='y', bins=50); ax[2].set_xlabel('Paragraph words');  ax[2].set_ylabel('Counts')
    plt.show()
```

```python
show_dataset_statistics(path='../datasets/en_wiki.txt', max_lines=1000)
```
